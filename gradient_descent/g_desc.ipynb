```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples=4,n_features=1, n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

### let's apply OLS

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
# lets apply Gradient descent assuming slope is costant m=78.35
# and lets assume the starting value for intercept b is 100
```

```python
y_pred =  ((78.35 * X)+100).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X), color='red')
plt.plot(X,y_pred, color='green')
plt.show()
```

```python
m = 78.35
b = 100
loss_slope = -2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
learning_rate = 0.1
step_size = loss_slope*learning_rate
step_size
```

```python
b = b-step_size
b
```

```python
y_pred1=  ((78.35 * X)+b).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='black',label='b = {}'.format(b))
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
# iteration 2
loss_slope = -2*np.sum(y-m*X.ravel()-b)
step_size = loss_slope*learning_rate
step_size
```

```python
b = b-step_size
b
```

```python
y_pred2=  ((78.35 * X)+b).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='darkgreen',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='black',label='b = {}'.format(b))
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
# iteration 3
loss_slope = -2*np.sum(y-m*X.ravel()-b)
step_size = loss_slope*learning_rate
step_size
```

```python
b = b-step_size
b
```

```python
y_pred3=  ((78.35 * X)+b)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='darkgreen',label='b = {}'.format(b))
plt.plot(X,y_pred3,color='blue',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='black',label='b = {}'.format(b))
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
b= 100
m=78.35
lr=0.1
epochs = 5
plt.plot(X,reg.predict(X),color='red',label='OLS')
for i in range(epochs):
    loss_slope = -2 * np.sum(y-m*X.ravel()-b)
    b= b -(lr*loss_slope)
    y_pred = m*X+b
    plt.plot(X,y_pred)
plt.scatter(X,y)
plt.show()
```
